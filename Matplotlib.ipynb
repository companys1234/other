```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y,color='red',marker='o',linestyle='--')
plt.xlabel('лыв')
plt.ylabel('лы2в')
plt.title('кпавлщ')
plt.legend()
plt.show()
```

```python
plt.bar(['A', 'B', 'C'], [3, 7, 2])
```

```python
data = [1, 1, 2, 3, 3, 3, 4, 5]
plt.hist(data, bins=5)
```

```python
plt.pie([30, 20, 50], labels=['авпа', 'B', 'C'])
```

```python
plt.scatter(x, y, color='red', s=100)  # s — размер точек
```

```python
plt.grid(True, linestyle='--', alpha=1)  # alpha — прозрачность
```

```python
plt.xlim(0, 5)
plt.ylim(0, 40)
```

```python
plt.figure(figsize=(10, 5))  # ширина, высота
plt.plot(x, y)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
ax[0, 0].plot(x, y)
ax[0, 1].bar(['A', 'B'], [3, 5])
ax[1, 0].scatter(x, y)
ax[1, 1].hist(data)
plt.tight_layout()
```
